```python
#transform polygon into geojsonand extract the disasters dates
engine = create_engine(url_engine)

sql_query = f"SELECT event_id,country,name,geometry_validated from disasters WHERE event_id = 1000072"
sql_query_date = f"SELECT event_id,fromdate,geometry_validated,todate from disasters WHERE event_id = 1000072"
disasters = gpd.read_postgis(sql_query, engine, geom_col="geometry_validated")
disasters_date = gpd.read_postgis(sql_query_date, engine, geom_col="geometry_validated")

todate=str(disasters_date['todate'][0])
fromdate=str(disasters_date['fromdate'][0])

#geojson to be used for requests.post / get
geojson_polygon=json.loads(disasters.to_json(show_bbox=False))
```

```python
#define headers 
headers_api_get= {'x-api-key': api_key,'accept': 'application/json'}
headers_api_post= {'x-api-key': api_key,'accept': 'application/json','Content-Type': 'application/json'}
```

```python
# GET structure 
requests.get('https://api.bazze.io/v1/users?wait=false&dryrun=false&run_large_query=false&limit=10&from_date=2023-06-01&to_date=2023-06-05&wifi_ssid=TP-LINK&country=RU&visited_country=RU' 
            , headers=headers_api_post).json()
```

```python
#function to use the POST structure 
def bazze_post(url, fromdate,todate,headers_api, json_application):
    query_id_json=requests.post(url+f'?wait=false&dryrun=false&run_large_query=false&limit=ALL&from_date={fromdate}&to_date={todate}' 
              , headers = headers_api_post
              , json = json_application).json()
    return query_id_json
```

```python
#function to get the results out of a query id
def get_results(query_id_json):
    try:
        results=pandas.read_csv(requests.post('https://api.bazze.io/v1/results' 
              , headers = headers_api_post
              , json = query_id_json).json()['csv'])
    except: 
        results='Query still running'
    return results
```

# GET THE RECORDS FROM THE USER LIST


```python
# spit the total user list in smaller lists
#do not go above 1000
list_length=500

#get the full list 
full_list_users=df_users_geofence['advertising_id'].tolist()

#split the full list in list of list of len list_length
list_list_users=[full_list_users[i * list_length:(i + 1) * list_length] for i in range((len(full_list_users) + list_length - 1) // list_length )] 
```

```python
fromdate_records=str(disasters_date['fromdate'][0]+datetime.timedelta(from_date_time_delta_records))
todate_records=str(disasters_date['todate'][0]+datetime.timedelta(to_date_time_delta_records))
```

```python
records_query_id=[]
i=0
# for each list within the list of list
for list_users in list_list_users:
    # get the query id from the records query
    
    print(i)
    #wrangling to have the list of users in the proper format - to be improved
    list_users=str(list_users)
    list_users=list_users.replace("'",'"')
    user_list_input_json='{"advertising_id":'+list_users+'}'
    user_list_input_json=json.loads(user_list_input_json)
    
    #get the query id via records/multiple call 
    query_id=bazze_post('https://api.bazze.io/v1/records/multiple',fromdate_records,todate_records,headers_api_post, user_list_input_json)
    records_query_id.append(query_id)
    i=i+1
```

```python
# extract and concatenate the output from the query_id
df_results=pd.DataFrame()
list_query_completed=[]
list_query_running=[]

i=0
# for each query id from the list 
for query_id in records_query_id:
    print(query_id)
    result=get_results(query_id)
    
    if type(result)==pandas.core.frame.DataFrame:
        list_query_completed.append(query_id)
        if i==0:
            df_results=result
        else:
            df_results=df_results.append(result)
        i=i+1
        
    else: 
        list_query_running.append(query_id)

print(f'there are {len(list_query_running)} queries still running')
```

# Calculate the flow matrix for each day


```python
# create 2 data flows df that will be joined to create the flow matrix
df_flow_1=df_records_geofence[['date','hash_user_id','latitude','longitude']]
df_flow_1['date_next']=df_flow_1['date']+datetime.timedelta(1)

df_flow_2=df_records_geofence[['date','hash_user_id','latitude','longitude']]
```

```python
#df_flow_1.join(df_flow_2, on , how='left', lsuffix='day_i', rsuffix='day_i+1', sort=False, validate=None)
```

```python
df_flow_matrix=pd.merge(df_flow_1, df_flow_2,  how='left'
                        , left_on=['date_next','hash_user_id'], right_on = ['date','hash_user_id']
                        , suffixes=('', '_nextday')
                       )
```

```python
df_flow_matrix=df_flow_matrix[['date','date_next','hash_user_id','latitude','longitude','latitude_nextday','longitude_nextday']]
df_flow_matrix
```

```python
# fill missing next day lat/lon with previous day lat lon 
# when there is no data for the next day we assume there was no movement

df_flow_matrix.latitude_nextday=df_flow_matrix.latitude_nextday.combine_first(df_flow_matrix.latitude)
df_flow_matrix.longitude_nextday=df_flow_matrix.longitude_nextday.combine_first(df_flow_matrix.longitude)
df_flow_matrix
```
